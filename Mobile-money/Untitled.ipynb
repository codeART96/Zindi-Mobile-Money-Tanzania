```python
#USING mwm_df to ROUGHLY SHOW THE RELATIONSHIP BETWEEN MALES AGES WITH AND WITHOUT MOBILE MONEY AND ARE MARRIED
married = len(mwm_df[mwm_df.Q3 == 1])
divorced = len(mwm_df[mwm_df.Q3 == 2])
widowed = len(mwm_df[mwm_df.Q3 == 3])
single = len(mwm_df[mwm_df.Q3 == 4])

marital_status = [married, divorced, widowed, single]
```

```python
import numpy as np
import matplotlib.pyplot as plt

 
#names of the bars
bars = ('married', 'divorced', 'widowed', 'single')
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, marital_status)
 
#Label
plt.ylabel('Number Of Recorded Males', color='orange', fontsize=12)

# names on the x-axis
plt.xticks(y_pos, bars)
#plt.yticks(color='orange')
plt.title('Marital Statuses Groups', color='orange', fontsize=20)

plt.show()
```
